## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
N=int(input("Entrer the number"))

if N<=1:
    print("Not prime")
else:
    is_prime=True
    for i in range(2,N):
        if N%i==0:
            is_prime=False
            break
    if is_prime==True:
            print("Prime")
    else:
            print("not prime")

```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
n = int(input("Enter a number: "))

for num in range(2, n + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime == True:
        print(num)
```

## Question 19 (Hard)

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:

123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
N = int(input("Enter a number: "))

for i in range(1, N + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
n = int(input("Enter a number: "))

sum_of_factors = 0

for i in range(1, n):
    if n % i == 0:
        sum_of_factors = sum_of_factors + i

if sum_of_factors == n:
    print("Perfect Number")
else:
    print("Not a Perfect Number")
```
