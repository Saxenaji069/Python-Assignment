## Problem 2: Leap Year

**Question:** Check if a given year is a leap year.

**Approach:** Apply the rules: divisible by 4, but centuries must be divisible by 400.

Rules: Year divisible by 4 is leap, but if divisible by 100 it must also be divisible by 400.

```python
year=int(input("Enter year"))

if (year%4==0 and year%100!=0) or (year%400==0):
    print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

## Problem 3: Grade Bands

**Question:** Print grade category based on marks: Excellent, Good, Average, Fail.

**Approach:** Build multiple conditions using elif to check ranges.

Take marks (0-100) and print:
- 90–100 → Excellent
- 75–89 → Good
- 50–74 → Average
- else → Fail

```python
marks=int(input("enter marks"))
if marks>90 and marks==100:
    print("Excellent") 
elif marks>=75 and marks<=89:
    print("Good")
elif marks>=50 and marks<=74:
    print("Average")
else:
    print("Fail")
```

## Problem 7: Delivery Charge System

**Question:** Calculate final delivery charge based on prime membership, amount, and distance.

**Approach:** Start with prime-check, then inside it apply nested if blocks.

Inputs:
- order_amount (int)
- is_prime_member (yes/no)
- distance_km (float)

Rules:
- If prime member → Free delivery
- Else:
  - If amount ≥ 500:
    - If distance ≤ 5 km → Free
    - Else → ₹20 extra
  - Else:
    - If distance ≤ 5 km → ₹30
    - Else → ₹50

```python
order_amount=int(input("Enter amount"))
is_prime_member=str(input("Enter if prime member (yes/no)"))
distance_km=float(input("Enter distance in km"))

if is_prime_member=="yes":
    print("free delivery")
else:
    if order_amount>=500:
        if distance_km<=5:
            print("Free delivery")
        else:
             print("Delivery with 20 rupees extra charge")
    else:
        if distance_km<=5:
            print("Delivery with 30 rupees extra charge")
        else:
            print("Delivery with 50 rupees extra charge")   
```

## Problem 8: ATM Withdrawal Simulation

**Question:** Simulate ATM rules: multiples of 100, sufficient balance, manager approval.

**Approach:** Check errors first, then nested conditions for amount > 10000.

Inputs:
- balance
- amount

Logic:
- If amount % 100 != 0 → reject (ATM works with ₹100 multiples)
- If amount > balance → reject
- Else:
  - if amount > 10000 → ask for manager approval
  - else → withdraw success

```python
balance=int(input("Enter balance"))
amount=int(input("Enter amount to withdraw"))

if amount%100!=0:
    print("Rejected: Amount must be a multiple of 100")
elif amount>balance:
        print("Rejected: Insufficient funds")
else:
      if amount>10000:
            print("ask for manager approval")
      else:
            balance-=amount
            print("Withdrawal successful.")
            
```
