# Q15 - Input: Circle calculations

- Circumference = `2 * 3.14159 * radius`

```python
raduis=float(input("Enter the radius"))
Circumference=2*3,14159*raduis
Area=3.14159*raduis**2
print("Circumference:",Circumference,"Area",Area)
```

# Q16 - Input: Convert total seconds to hours, minutes, seconds
